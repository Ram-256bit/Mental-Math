```python
import pandas as pd
import numpy as np
```

```python
x = np.arange(1000, 10000)
```

```python
y = np.arange(1, 100)
```

```python
all_numbers = np.transpose([np.tile(x, len(y)), np.repeat(y, len(x))])
df = pd.DataFrame(all_numbers)
# df=df[df[0] < df[1]]
df = df.sample(frac=1)[:20000].reset_index(drop=True)
```

```python
# np.random.shuffle(x)
# np.random.shuffle(y)

# df = pd.DataFrame()
# df[0] = x[:10000] 
# df[1] = y[:10000]
```

```python
df['Question'] = df[0].astype(str) + " / " + df[1].astype(str)
```

```python
df['Answer'] = df[0] / df[1]
```

```python
df = df[['Question', 'Answer']]
```

```python
df.sample(frac=1).to_csv("four_two_division.csv", index=None, header=None)
```
