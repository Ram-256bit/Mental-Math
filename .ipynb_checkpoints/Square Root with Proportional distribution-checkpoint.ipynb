```python
import pandas as pd
import numpy as np
```

```python
x = np.arange(100, 1000)
y = np.arange(1000, 10000)
z = np.arange(10000, 100000)
a = z = np.arange(100000, 1000000)

np.random.shuffle(x)
np.random.shuffle(y)
np.random.shuffle(z)
np.random.shuffle(a)
```

```python
sqrt = np.concatenate((x[:3000], y[:3000], z[:3000], a[:3000]))
np.random.shuffle(sqrt)
```

```python
df = pd.DataFrame(columns=['Number', 'Square Root'])
```

```python
df['Number'] = sqrt
```

```python
df['Square Root'] = np.around(np.sqrt(df['Number']), 2)
```

```python
df = df[['Number', 'Square Root']]
```

```python
df.sample(frac=1).to_csv("square_root.csv", index=None, header=None)
```
