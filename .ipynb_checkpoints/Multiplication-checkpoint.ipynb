```python
import pandas as pd
import numpy as np
```

```python
x = np.arange(100, 100000)
```

```python
y = np.arange(100, 100000)
```

```python
# all_numbers = np.transpose([np.tile(x, len(y)), np.repeat(y, len(x))])
# df = pd.DataFrame(all_numbers)
```

```python
np.random.shuffle(x)
np.random.shuffle(y)

df = pd.DataFrame()
df[0] = x[:10000] 
df[1] = y[:10000]
```

```python
df['Question'] = df[0].astype(str) + " × " + df[1].astype(str)
```

```python
df['Answer'] = df[0] * df[1]
```

```python
df = df[['Question', 'Answer']]
```

```python
df.sample(frac=1).to_csv("big_multiplication.csv", index=None, header=None)
```
